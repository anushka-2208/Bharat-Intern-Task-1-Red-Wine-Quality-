### Importing the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing the Red Wine data

```python
wine_dataset= pd.read_csv("/kaggle/input/red-wine-quality-cortez-et-al-2009/winequality-red.csv")
wine_dataset
```

### Analyzing through Visualization

```python
# number of values for each quality
sns.catplot(x='quality', data = wine_dataset, kind = 'count')
```

### Splitting the independent and the dependent features

```python
# separate the data and Label
X = wine_dataset.drop('quality',axis=1)
```

```python
Y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
```

### Splitting the training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

### Random Forest Algorithim

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

### Accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy : ', test_data_accuracy)
```
